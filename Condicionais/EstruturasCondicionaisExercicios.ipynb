2- Faça um algoritmo que peça ao usuário um valor de salário e um valor de gasto
mensal. Se o salário for menor que o gasto mensal, apresentar saldo negativo,
senão saldo positivo.

```python
# Seu código aqui
sal = float(input("Salário: "))
gast_men = float(input("Qual é o gasto mensal: "))
saldo = sal - gast_men
if gast_men > sal:
  print("Seu saldo é negativo R$", saldo)
elif gast_men == sal:
  print("Seu saldo é nulo")
else:
  print("Seu saldo é positivo R$", saldo)
```

4- Faça um algoritmo que leia o nome de um aluno e as 2 notas dos 4 bimestres.
No final informar o nome do aluno e se a média das 2 notas for maior ou igual a 7, apresentar a mensagem Aprovado, senão apresentar mensagem Reprovado.

```python
# Seu código aqui
name = input("Nome do Aluno: ")
n1 = float(input("1ª nota do 1º bimestre: "))
n2 = float(input("2ª nota do 1º bimestre: "))
med_1 = (n1 + n2) / 2
print(name, "sua média no 1º bimestre foi de ", med_1)
n1 = float(input("1ª nota do 2º bimestre: "))
n2 = float(input("2ª nota do 2º bimestre: "))
med_2 = (n1 + n2) / 2
print(name, "sua média no 2º bimestre foi de ", med_2)
n1 = float(input("1ª nota do 3º bimestre: "))
n2 = float(input("2ª nota do 3º bimestre: "))
med_3 = (n1 + n2) / 2
print(name, "sua média no 3º bimestre foi de ", med_3)
n1 = float(input("1ª nota do 4º bimestre: "))
n2 = float(input("2ª nota do 4º bimestre: "))
med_4 = (n1 + n2) / 2
print(name, "sua média no 4º bimestre foi de ", med_4)
med_g = (med_1 + med_2 + med_3 + med_4) / 4
if med_g >= 7:
  print(name, "você foi aprovado/a com a media de ", med_g )
else:
  print(name, "você foi reprovado/a com a media de ", med_g )
```

12 -Fazer um programa onde o lojista entra com o preço do produto e o programa traz opções de escolha ao cliente antes de efetuar o pagamento :

O  logista vai avaliar o total da compra e após isso escolher destre essas opções, de acordo com o que o cliente optar.

* a) 10% para pagamento a vista
* b) Prestação de 2x sem juros, se o valor for maior que 200 reais
* c) Prestação de 3x sem juros, se o valor for maior que 300 reais
* d) Prestação de 4x sem juros, se o valor for maior que 400 reais
* e) Prestação de 5x sem juros, se valores superiores a 500 reais.
* c) Prestação de 10x com juros compostos de 2% ao mês

```python
print("Lojas do Marcos")
preco = float(input("Qual é o preço: "))

print("Formas de pagamento")
print("1°) 10% para pagamento a vista")
if preco > 200:
  print("2°) Prestação de 2x sem juros, se o valor for maior que 200 reais")
elif preco > 300:
  print("3°) Prestação de 3x sem juros, se o valor for maior que 300 reais")
elif preco > 400:
  print("4°) Prestação de 4x sem juros, se o valor for maior que 400 reais")
elif preco > 500:
  print("5°) Prestação de 5x sem juros, se valores superiores a 500 reais.")

print("6°) Prestação de 10x com juros compostos de 2% ao mês")

opc = int(input("Qual será a forma de pagamento? "))
if opc == 1:
  print("(1°) 10% para pagamento a vista) opção selecionada")
  pag = preco - (preco * 10 / 100)
  print("Preço com desconto R$", pag)
elif opc == 1 and preco > 200:
  print("(2°) Prestação de 2x sem juros, se o valor for maior que 200 reais) opção selecionada")
  pag = preco/2
  print("Preço das parcelas R$", pag,)
elif opc == 2 and preco > 300:
  print("(3°) Prestação de 3x sem juros, se o valor for maior que 300 reais) opção selecionada")
  pag = preco/3
  print("Preço das parcelas R$", pag)
elif opc == 4 and preco > 400:
  print("(4°) Prestação de 4x sem juros, se o valor for maior que 400 reais) opção selecionada")
  pag = preco/4
  print("Preço das parcelas R$", pag)
elif opc == 5 and preco > 500:
  print("(5°) Prestação de 5x sem juros, se valores superiores a 500 reais) opção selecionada")
  pag = preco/5
  print("Preço das parcelas R$", pag)
elif opc == 6:
  print("(6°) Prestação de 10x com juros compostos de 2% ao mês) opção selecionada")
  pag = (preco * (1 + (2 / 100)) ** 10)
  print("Preço final a ser pago pelo produto é de R$", pag)
else:
  print("Opção Inválida")
```
