2- Faa um algoritmo que pea ao usurio um valor de salrio e um valor de gasto
mensal. Se o salrio for menor que o gasto mensal, apresentar saldo negativo,
seno saldo positivo.

```python
# Seu cdigo aqui
sal = float(input("Salrio: "))
gast_men = float(input("Qual  o gasto mensal: "))
saldo = sal - gast_men
if gast_men > sal:
  print("Seu saldo  negativo R$", saldo)
elif gast_men == sal:
  print("Seu saldo  nulo")
else:
  print("Seu saldo  positivo R$", saldo)
```

12 -Fazer um programa onde o lojista entra com o preo do produto e o programa traz opes de escolha ao cliente antes de efetuar o pagamento :

O  logista vai avaliar o total da compra e aps isso escolher destre essas opes, de acordo com o que o cliente optar.

* a) 10% para pagamento a vista
* b) Prestao de 2x sem juros, se o valor for maior que 200 reais
* c) Prestao de 3x sem juros, se o valor for maior que 300 reais
* d) Prestao de 4x sem juros, se o valor for maior que 400 reais
* e) Prestao de 5x sem juros, se valores superiores a 500 reais.
* c) Prestao de 10x com juros compostos de 2% ao ms

```python
print("Lojas do Marcos")
preco = float(input("Qual  o preo: "))

print("Formas de pagamento")
print("1) 10% para pagamento a vista")
if preco > 200:
  print("2) Prestao de 2x sem juros, se o valor for maior que 200 reais")
elif preco > 300:
  print("3) Prestao de 3x sem juros, se o valor for maior que 300 reais")
elif preco > 400:
  print("4) Prestao de 4x sem juros, se o valor for maior que 400 reais")
elif preco > 500:
  print("5) Prestao de 5x sem juros, se valores superiores a 500 reais.")

print("6) Prestao de 10x com juros compostos de 2% ao ms")

opc = int(input("Qual ser a forma de pagamento? "))
if opc == 1:
  print("(1) 10% para pagamento a vista) opo selecionada")
  pag = preco - (preco * 10 / 100)
  print("Preo com desconto R$", pag)
elif opc == 1 and preco > 200:
  print("(2) Prestao de 2x sem juros, se o valor for maior que 200 reais) opo selecionada")
  pag = preco/2
  print("Preo das parcelas R$", pag,)
elif opc == 2 and preco > 300:
  print("(3) Prestao de 3x sem juros, se o valor for maior que 300 reais) opo selecionada")
  pag = preco/3
  print("Preo das parcelas R$", pag)
elif opc == 4 and preco > 400:
  print("(4) Prestao de 4x sem juros, se o valor for maior que 400 reais) opo selecionada")
  pag = preco/4
  print("Preo das parcelas R$", pag)
elif opc == 5 and preco > 500:
  print("(5) Prestao de 5x sem juros, se valores superiores a 500 reais) opo selecionada")
  pag = preco/5
  print("Preo das parcelas R$", pag)
elif opc == 6:
  print("(6) Prestao de 10x com juros compostos de 2% ao ms) opo selecionada")
  pag = (preco * (1 + (2 / 100)) ** 10)
  print("Preo final a ser pago pelo produto  de R$", pag)
else:
  print("Opo Invlida")
```
