1) Identifique os dados de entrada, processamento e saída no algoritmo abaixo:

Receba Valor da peça (VALOR) Receba Quantidade de peças (QTDE) Entrada

Calcule o valor total da peça (VT) (VT = QTDE * VALOR) P

```python
print("Este programa calculará o preço a ser pago para certa quantidade de peças")
#Entrada
VALOR = float(input("Digite o valor da peça "))
QTDE = float(input("Digite a quantidade de peças "))

#Processamento
VT =  QTDE * VALOR

#Saída
print("O valor total a ser pago é ", VT)
```
